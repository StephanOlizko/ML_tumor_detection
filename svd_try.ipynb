```python
#Загружаем изображение в формате jpg в виде матрицы пикселей
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('data\jpg\image_mutation_anonym\Person1_8.jpg')

#Преобразуем изображение в матрицу пикселей
img = np.array(img)

#Выведем размеры изображения
print(img.shape)

#Выведем изображение черно-белое
plt.imshow(img, cmap='gray')
plt.show()
```

```python
#Выполним svd-разложение изображения
U, s, V = np.linalg.svd(img, full_matrices=True)

#Рандомный вектор размера s
rm = np.random.rand(s.shape[0]) * 10

s = s + rm

#Восстановим изображение
img_restored = U @ np.diag(s) @ V

plt.imshow(img, cmap='gray')
plt.show()

#Выведем изображение черно-белое
plt.imshow(img_restored, cmap='gray')
plt.show()

plt.imshow(img_restored-img, cmap='gray')
plt.show()
```
